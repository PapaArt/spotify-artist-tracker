```python
import pandas as pd
import networkx as nx
import json
```

```python
data = pd.read_csv("../data/pop_world.csv")
```

```python
data
```

```python
G = nx.DiGraph()
```

```python
for _, row in data.iterrows():
    main_artist = row["main_artist"]
    main_artist_genres = eval(row["genres"])
    conjunto1 = set(main_artist_genres)
    relations = eval(row["relations"])
    
    G.add_node(main_artist)
    
    for relation in relations:
        artist_relation = relation["name"] 
        artist_genres = relation["genres"].split(", ")
        conjunto2 = set(artist_genres) 
        
        weight = len(conjunto1.intersection(conjunto2)) + 1
        
        G.add_edge(main_artist, artist_relation, weight=weight)        
```

```python
nx.write_gml(G, "../resources/gephi/pop_world.gml")
```
